반복문과 조건문 함께 사용


```python
scores = [42, 60, 77, 89, 20]
for score in scores:
    if score >= 60:
        print(f'{score}점은 합격입니다.')
    else:
        print(f'{score}점은 불합격입니다.')   # option + shift + 방향키 (↑, ↓) : 여러 줄 선택
```

```python
scores = [65, 82, 91, 100, 58, 77]

count = 0
for a in scores:
    if a >= 70:
        count += 1
print(f'합격자는 {count}명입니다.')            #f-string : 문자열 포매팅 방법 중 하나, 변수나 표현식을 문자열 안에 직접 삽입할 수 있다. f'문자열 {변수}' 형태로 사용한다.
```
